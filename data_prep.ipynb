```python
import os
import kagglehub
```

```python
# Dataset location
dataset_loc = os.path.join(os.getcwd(), "dataset\\samples")
dataset_loc
```

```python
# I used kaggle API and manually moved files across into the dataset_loc location

path = kagglehub.dataset_download(
    "fournierp/captcha-version-2-images"
)

print("Dataset downloaded to:", path)
```

```python
## General check
count = 0
min_len, max_len = 0, 0
count_png = 0
count_jpg = 0
for i in os.listdir(dataset_loc):
    count += 1
    min_len = max(min_len, len(i))
    max_len = min(min_len, len(i))
    if i[-3:] == "png":
        count_png += 1
    if i[-3:] == "jpg":
        count_jpg += 1

print(count)
print(min_len)
print(max_len)
print(count_png)
print(count_jpg) 
```

Found 30 jpg files, will change them to png for simplicity.

Loop through -> open files -> change to png


```python
from PIL import Image
```

```python
for file in os.listdir(dataset_loc):
    if file[-3:] == "jpg":
        # Open image
        img_path = os.path.join(dataset_loc, file)
        img = Image.open(img_path)

        new_filename = os.path.splitext(file)[0] + ".png"
        new_path = os.path.join(dataset_loc, new_filename)

        img.save(new_path)

        # Delete old
        os.remove(img_path)

        print(f"Converted {file} → {new_filename}")
```

Re-run the previous check to see if all files have converted to png

```python
count = 0
min_len, max_len = 0, 0
count_png = 0
count_jpg = 0
for i in os.listdir(dataset_loc):
    count += 1
    min_len = max(min_len, len(i))
    max_len = min(min_len, len(i))
    if i[-3:] == "png":
        count_png += 1
    if i[-3:] == "jpg":
        count_jpg += 1

print(count)
print(min_len)
print(max_len)
print(count_png)
print(count_jpg)
```

Check if there is variation in size or type

```python
image_size_miss = 0
image_mode_miss = 0
exp_size = (200, 50)


for i in os.listdir(dataset_loc):
    path = os.path.join(dataset_loc, i)
    img = Image.open(path)
    if img.size != exp_size:
        image_size_miss += 1
    if img.mode != "L":
        image_mode_miss += 1

print(image_size_miss)
print(image_mode_miss)
```

Size and mode all match! We are good to go!
